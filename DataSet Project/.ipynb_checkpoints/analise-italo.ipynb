## Normalização dos Dados

A normalização é uma fase crucial do nosso projeto. Os dados que temos contêm muitas informações fora da base binária, o que pode resultar em grandezas fora de escala ao serem comparadas em gráficos.

Para resolver isso, normalizamos nossos dados, definindo um intervalo de 0 a 1. Esse processo garante que os dados mantenham seus valores originais, mas em uma escala comum a todos, facilitando comparações e análises visuais.
