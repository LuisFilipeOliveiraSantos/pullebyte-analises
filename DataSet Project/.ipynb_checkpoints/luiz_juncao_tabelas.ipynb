# Salvando os dados descritivos aplicando One-hot encoding
Pegamos os dados de texto contidos no campo descrição e dummerizamos em diversas colunas

```python
unique_values = set()

# Itera sobre a coluna 'description' e extrai os valores únicos
for description in df_game_events['description']:
  if pd.notna(description):
    # Remove números e pontos no início da descrição
    description = description.split('. ', 1)[-1]
    values = description.split(', ')
    unique_values.update(values)

# Cria novas colunas no DataFrame para cada valor único
for value in unique_values:
  df_game_events[value.replace(' ', '_')] = df_game_events['description'].apply(lambda x: 1 if (pd.notna(x) and value in x.split('. ', 1)[-1]) else 0)

# Remove a coluna original 'description' se desejar
df_game_events = df_game_events.drop('description', axis=1)

# Save the DataFrame to a parquet file
df_game_events.to_parquet("Q:/Universidade/PISI3/pullebyte-analises/DataSet Project/transfermarkrt-dados-clean/dummy_data/game_events_description_dummy.parquet")
```

# Junção dos datasets
Aqui realizamos a junção dos dataset que jugamos relevantes para a analise

```python
df_games = pd.read_parquet("Q:/Universidade/PISI3/pullebyte-analises/DataSet Project/transfermarkrt-dados-clean/games.parquet")
```

```python
# seleciona colunas requeridas
df_games = df_games[['game_id', 'competition_id', 'season','round','date','home_club_id','away_club_id', 'home_club_goals','away_club_goals']]
# df_games
```

Junção do games & game_events_dummy

```python
merge_games_events_dummy = df_games.merge(df_game_events_dummy_filtred, on='game_id')
# merge_games_events_dummy
```

```python
df_players = pd.read_parquet("Q:/Universidade/PISI3/pullebyte-analises/DataSet Project/transfermarkrt-dados-clean/players.parquet")
# df_players.columns.to_list()
```

```python
df_players = df_players[['player_id', 'current_club_id', 'name', 'first_name', 'last_name','height_in_cm', 'date_of_birth', 'foot', 'position','sub_position', 'market_value_in_eur', 'highest_market_value_in_eur']]
# df_players
```

Junção entre merge_games_events_dummy & players

```python
merge_games_events_dummy_players = merge_games_events_dummy.merge(df_players, on='player_id')
# merge_games_events_dummy_players
```

```python
df_clubs = pd.read_parquet("Q:/Universidade/PISI3/pullebyte-analises/DataSet Project/transfermarkrt-dados-clean/clubs.parquet")

# Select the columns you want to keep from the clubs dataframe
df_clubs = df_clubs[['club_id', 'name' ]]

# Join the clubs dataframe with the merge_games_events_dummy_players dataframe
df_merged = merge_games_events_dummy_players.merge(df_clubs, left_on=['home_club_id'], right_on=['club_id'], how='left')
df_completo = df_merged.merge(df_clubs, left_on=['away_club_id'], right_on=['club_id'], how='left', suffixes=('_home', '_away'))

df_completo
```

# Tratamento dos dados recebido
Fazendo a padronização dos dados do tabelão

```python
from sklearn.preprocessing import StandardScaler

# Select the columns to standardize
columns_to_standardize = ['minute', 'height_in_cm', 'market_value_in_eur', 'highest_market_value_in_eur']

# Create a StandardScaler object
scaler = StandardScaler()

# Standardize the selected columns
df_completo[columns_to_standardize] = scaler.fit_transform(df_completo[columns_to_standardize])
df_completo
```
